# **Cancer Prediction**

Dataset Information:

Target Variable (y):
- Diagnosis (M = malignant, B = benign)

Ten features (X) are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter)
2. texture (standard deviation of gray-scale values)
3. perimeter
4. area
5. smoothness (local variation in radius lengths)
6.  compactness (perimeter^2 / area - 1.0)
7. concavity (severity of concave portions of the contour)
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension (coastline approximation - 1)

For each characteristic three measures are given:

a. Mean

b. Standard error

c. Largest/ Worst

```python
# Step 1 : import library
import pandas as pd
```

```python
# Step 2 : import data
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
# Step 3 : define target (y) and features (X)
```

```python
cancer.columns
```

```python
y = cancer['diagnosis']
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

# The model achieved high accuracy, indicating strong predictive performance.


```python
print(classification_report(y_test,y_pred))
```

```python
# Model Evaluation Improvement
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion Matrix Visualization
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.show()
```

```python
# 🔥 Feature Importance (Top-Level Improvement)
import pandas as pd
import matplotlib.pyplot as plt

try:
    importance = model.feature_importances_
    features = X.columns

    feat_df = pd.DataFrame({'Feature': features, 'Importance': importance})
    feat_df = feat_df.sort_values(by='Importance', ascending=False)

    print("Top Important Features:")
    print(feat_df.head(10))

    # Plot
    plt.figure()
    plt.barh(feat_df['Feature'][:10], feat_df['Importance'][:10])
    plt.gca().invert_yaxis()
    plt.title("Top 10 Feature Importances")
    plt.show()

except:
    print("Feature importance not available for this model. Use RandomForest for this feature.")
```
